```python
#visualization time series
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
#%matplotlib notebook # for interactive graph inside notebook
mac=pd.read_csv('mcdonalds.csv',index_col='Date',parse_dates=True)

mac.head()
mac['Adj. Close'].plot(xlim=['2007-01-01','2009-01-01'],ylim=(20,50),ls='--',c='red')
plt.show()
mac['Adj. Volume'].plot(figsize=(12,3))

import matplotlib.dates as dates
idx=mac.loc['2007-01-01':'2007-05-01'].index
stock=mac.loc['2007-01-01':'2007-05-01']['Adj. Close']

fig,ax=plt.subplots()
ax.plot_date(idx,stock,'-')
#LOCATION
ax.xaxis.set_major_locator(dates.MonthLocator())
ax.xaxis.set_major_formatter(dates.DateFormatter('\n\n%b-%Y'))

#b-for half month name
#B-for full month name
#\n -push out to input format data
ax.xaxis.set_minor_locator(dates.WeekdayLocator(byweekday=0))#change 0to 1,2,3,4,, for mon tue wed thur fri sat sun
ax.xaxis.set_minor_formatter(dates.DateFormatter('%d'))

#d=date
#a =weekname

fig.autofmt_xdate() #twist the date to look nicer
ax.yaxis.grid(True)
ax.xaxis.grid(True)
plt.tight_layout()





```

```python
#DateTime Index
#Time Resampling
#Time Shifts
#Rolling and Expanding

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
from datetime import datetime
year=2017
month=1
day=2
hour=10
minute=15
second=15
date=datetime(year,month,day,hour,minute,second)
print(date.day)
listd=[(datetime(2015,1,1)),(datetime(2016,2,3))]
dfdate=pd.DatetimeIndex(listd)
print(dfdate)

#time resampling

df=pd.read_csv('walmart_stock.csv',index_col='Date',parse_dates=True)
print(df.index)
df.head()
#another way of converting string to datetime format but remove index_col and parse_dates attribute
#df['Date']=pd.to_datetime(df['Date'])
#print(type(df['Date']))
#df.set_index('Date',inplace=True)
#print(df)

df.resample(rule='A').mean()
# A- yearly
#Q-quartly resampling
#BQ -business quartly
#more help see table on github
df.resample(rule='A').max()
def first_day(entry):
    return entry[0]
df.resample('A').apply(first_day)
df['Close'].resample('M').mean().plot(kind='bar',figsize=(16,6))
#A-Yearly graph,M-for Monthly graph



#Time shift

x=pd.read_csv('walmart_stock.csv',index_col='Date',parse_dates=True)
x.shift(periods=1).head()
# push the data one row below to shift backward put -1
print(x)

# all the day of jan share the last day value for year use y

df.tshift(freq='M').head()



```

```python
#Roalling and expanding

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv('walmart_stock.csv',index_col='Date',parse_dates=True)
df['Open'].plot(figsize=(16,6))
# getting the average of the  week 
df.rolling(7).mean().head(14)
df['Open'].plot()
df.rolling(window=90).mean()['Close'].plot(figsize=(16,6))

df['Close 30 Day MA']=df['Close'].rolling(window=30).mean()
#creating a legend
df[['Close 30 Day MA','Close']].plot(figsize=(16,6))


#expanding basic mean from beginning of the time series to the current time series

df['Close'].expanding().mean().plot(figsize=(16,6))
#Bollinger Bands ti consist of Close,close of 20 day mean,Upper,Lower

#Closing prise 20 MA(moving average)
df['Close: 20 Day Mean']=df['Close'].rolling(20).mean()
#Upper =20MA + 2*std(20)
df['Upper']=df['Close: 20 Day Mean']+2*(df['Close'].rolling(20).std())

#Lower=20MA-2*std(20)
df['Lower']=df['Close: 20 Day Mean']-2*(df['Close'].rolling(20).std())

#close
df[['Close','Close: 20 Day Mean','Upper','Lower']].plot(figsize=(16,6))



```
