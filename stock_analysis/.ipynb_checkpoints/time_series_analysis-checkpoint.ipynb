#ARIMA(Auto regressive integrated Moving Average) (not work very well with stock market)
#-Non-seasonal ARIMA Model
#-Seasonal ARIMA model
#basic need 
#-mean need to be constant
#-variance need to be function of time(very very high up and dow)
#Covariance should not be function of time (quick change with time and slow change with time)
#use Augmented Dickey-Fuller to test if data is stationary or non stationary

#ACF and PACF
#an auto correlation plot(correlogram) shows the correlation of the
#series with itself ,lagged by x time unit.

#firstly copy the observation and delete the first obeservation of copy 1 and the last observationof copy 2
#now you have two series of length T-1 for which you calculated a correlation coefficient.This is the value of the vertical axis at X=1 in your plots.
#type1: Gradual Decline
#type2: sharp Drop off
#if the auto correlation plot shows negative auto correlation at the first lag then it suggest using MA(Moving average)
#if auto correlation plot show positive auto correlation at the first lag then it suggests using AR(ARIMA)

#p:The number of lag observations included in the model
#d:The number of times s that the raw observations are differenced
#q: the size of the moving average window also called the order of moving average
#identification of an AR model is often best done with the PACF.
#identification of an MA model id often best done with the ACF rather than the PACF







```python
#ARIMS
#Visualize the time Series Data
# Make the time series data stationary
#plot the Correlation and AutoCorrelation Chart
# Construct the ARIMA Model
# Use the model to make Predictions


import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
%matplotlib inline
df=pd.read_csv('monthly-milk-production-pounds-p.csv')
df.head()
#rename column
df.columns=['Month','Milk in Pounds per Cow']
df.drop(168,axis=0,inplace=True)
df['Month']=pd.to_datetime(df['Month'])
df.head()
df.set_index('Month',inplace=True)
df.index
df.describe().transpose()

#step2
df.plot()
time_series=df['Milk in Pounds per Cow']
type(time_series)
#mean
time_series.rolling(12).mean().plot(label='12 month rolling mean')
#standard deviation
time_series.rolling(12).std().plot(label='12 month rolling std')


time_series.plot()
from statsmodels.tsa.seasonal import seasonal_decompose
decompose=seasonal_decompose(time_series,freq=12)
fig=decompose.plot()
fig.set_size_inches(15,6)



```

```python


#testing the data is stationary or not
df.head()

from statsmodels.tsa.stattools import adfuller
result=adfuller(df['Milk in Pounds per Cow'])
print(result)
def adf_check(time_series):
    
    result=adfuller(time_series)
    print('Augmented Dicky-Fuller Test')
    labels=['ADF Test Statistic','p-value','# of lags','num of Observations used']
    for value,label in zip(result,labels):
        print(label+" : "+str(value))
    if result[1] <=0.05:
        print("Strong evidence againse null hypothesis")
        print("reject null hypothesis")
        print("Data has no unit root and isstationary")
    else:
        print("weak evidence against null hypothesis")
        print("Fail to reject null hypothesis")
        print("Data has a unit root,it is non-stationary")
        

adf_check(df['Milk in Pounds per Cow'])

df['First Difference']=df['Milk in Pounds per Cow']-df['Milk in Pounds per Cow'].shift(1)
df['First Difference'].plot()



```

```python
adf_check(df['First Difference'].dropna())
```

```python
df['Milk Second Difference']=df['First Difference']-df['First Difference'].shift()
adf_check(df['Milk Second Difference'].dropna())
```

```python
df['Seasonal Difference']=df['Milk in Pounds per Cow']-df['Milk in Pounds per Cow'].shift(12)
adf_check(df['Seasonal Difference'].dropna())

df['Seasonal Difference'].plot()



```

```python
df['Seasonal First Difference']=df['First Difference']-df['First Difference'].shift(12)
df['Seasonal First Difference'].plot()
adf_check(df['Seasonal First Difference'].dropna())

```

```python
#autocorelation plot and partial autocorelations plot

from statsmodels.graphics.tsaplots import plot_acf,plot_pacf

fig_first=plot_acf(df['First Difference'].dropna())


fig_Seasonal_first=plot_acf(df['Seasonal First Difference'].dropna())


```

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(df['Seasonal First Difference'].dropna())

result=plot_pacf(df['Seasonal First Difference'].dropna())

plot_acf(df['Seasonal First Difference'].dropna())
plot_pacf(df['Seasonal First Difference'].dropna() )




```

```python
from statsmodels.tsa.arima_model import ARIMA
model=sm.tsa.statespace.SARIMAX(df['Milk in Pounds per Cow'],order=(0,1,0),seasonal_order=(1,1,1,12))

results=model.fit()
print(results.summary())

results.resid.plot()
```

```python
results.resid.plot(kind='kde')
df['forecast']=results.predict(start=150,end=168)
df[['Milk in Pounds per Cow','forecast']].plot()

```

```python
from pandas.tseries.offsets import DateOffset
future_dates=[df.index[-1]+DateOffset(months=x) for x in range(1,24)]
print(future_dates)
 
future_df=pd.DataFrame(index=future_dates,columns=df.columns)
final_df=pd.concat([df,future_df])
final_df.head()

final_df['forecast']=results.predict(start=168,end=192)
final_df[['Milk in Pounds per Cow','forecast']].plot()

```
