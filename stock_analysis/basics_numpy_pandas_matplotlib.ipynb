```python
#Group by (grop rows based on column value)

import numpy as np
import pandas as pd

data={'Company':['pal','pal','tata','tata','google','google'],
     'person':['sanjeet','sagar','sachin','rohit','mohit','appoorv'],
      'Sales':[200,100,300,456,100,230]
     
     
     }
df=pd.DataFrame(data)

print(df)

group_company=df.groupby('Company')
print(group_company.sum().loc['pal'])
#you can use sum,average,std
pal=df.groupby('Company').describe().transpose()['pal']
print(pal )
```
